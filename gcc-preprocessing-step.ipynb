```python
import re
import os

# Specify the path to your input and output files
#input_file_path = './2023-April.txt'
#output_file_path = './2023-April-processed.txt'

def process_file(input_file_path):
# Read the content of the input file
    with open(input_file_path, 'r') as file:
        content = file.read()
  # Remove lines matching the specified format
    cleaned_content = re.sub(r'^From [a-zA-Z0–9._%+-@. ]+\d{2}:\d{2}:\d{2} \d{4}\n', '', content, flags=re.MULTILINE)
    # Extract the base filename without the path and extension
    base_filename = os.path.splitext(os.path.basename(input_file_path))[0]
    output_file_path = os.path.join('../2021/output', f'{base_filename}-processed.txt')
    print(output_file_path)
  # Write the cleaned content to the output file
    with open(output_file_path, 'w') as file:
        file.write(cleaned_content)
    
# Specify the input folder
input_folder = '../2021/input'

# Get a list of files in the input folder
file_list = [os.path.join(input_folder, file) for file in os.listdir(input_folder) if file.endswith('.txt')]

# Process each file
for input_file in file_list:
    process_file(input_file)
    
```

```python
import re
import pandas as pd
from email import policy
from email.parser import BytesParser

def process_files_df(input_file_path):
    print(input_file_path)
    # Read the content of the input file
    with open(input_file_path, 'rb') as file:
        content = file.read()

    # Use BytesParser from the email library to parse the content
    msgs = list(email.message_from_bytes(msg, policy=policy.default) for msg in re.split(b'\n(?=From:)', content))
    #print(msgs[0])
    # Initialize lists to store values
    from_list = []
    date_list = []
    subject_list = []
    message_id_list = []
    content_list = []
    cc_list = []
    to_list = []
   
    for msg in msgs:
        #print("MESSAGE:")
        #print(msg)
        from_line = msg.get('From')
        date_line = msg.get('Date')
        to_line = msg.get('To')
        cc_line = msg.get('Cc')
        subject_line = msg.get('Subject')
        reply_line = msg.get('In-Reply-To')
        ref_line = msg.get('References')
        message_id_line = msg.get('Message-ID')
        

        # Extract information from the "From" line
        from_match = re.match(r'^From: (.+?) \((.+?)\)$', from_line)
        if from_match:
            from_value = from_match.group(1).strip()
        else:
            from_value = from_line.strip()

        # Extract information from the "Date" line
        date_value = date_line.strip() if date_line else ''

        to_value = to_line.strip() if to_line else ''
        cc_value = cc_line.strip() if cc_line else ''
        # Extract information from the "Subject" line
        subject_value = subject_line.strip() if subject_line else ''

        # Extract information from the "Message-ID" line
        message_id_value = message_id_line.strip() if message_id_line else ''

        reply_value = reply_line.strip() if reply_line else ''
        ref_value = ref_line.strip() if ref_line else ''
        # Extract content
        content_value = msg.get_payload()

        # Append values to lists
        from_list.append(from_value)
        date_list.append(date_value)
        to_list.append(to_value)
        cc_list.append(cc_value)
        subject_list.append(subject_value)
        reply_list.append(reply_value)
        ref_list.append(ref_value)
        message_id_list.append(message_id_value)
        content_list.append(content_value)

    # Create a DataFrame
    df = pd.DataFrame({
        'From': from_list,
        'Date': date_list,
        'To': to_list,
        'CC': cc_list,
        'Subject': subject_list,
        'Message-ID': message_id_list,
        'Content': content_list
    })

    # Define the output file path
    output_file_path = re.sub(r'-processed.txt$', '-output.csv', input_file_path)
    print(output_file_path)
    # Write the processed content to the output file
    df.to_csv(output_file_path, index=False)    

# Specify the input folder
input_folder = '../2021/output'

# Get a list of files in the input folder
file_list = [os.path.join(input_folder, file) for file in os.listdir(input_folder) if file.endswith('.txt')]

# Process each file
for input_file in file_list:
    print(input_file)
    process_files_df(input_file)

print("Processing completed.")       
```

```python
import os
import pandas as pd

# Specify the input folder containing CSV files
input_folder = './output'

# Get a list of CSV files in the input folder
csv_files = [file for file in os.listdir(input_folder) if file.endswith('.csv')]

# Initialize an empty DataFrame to store the concatenated data
concatenated_df = pd.DataFrame()

# Iterate through each CSV file and concatenate the data
for csv_file in csv_files:
    file_path = os.path.join(input_folder, csv_file)
    df = pd.read_csv(file_path)
    concatenated_df = pd.concat([concatenated_df, df], ignore_index=True)

# Specify the output file path for the concatenated CSV
output_file_path = './output/gcc_2023_output.csv'

# Write the concatenated DataFrame to a new CSV file
concatenated_df.to_csv(output_file_path, index=False)

print("Concatenation completed.")
```
